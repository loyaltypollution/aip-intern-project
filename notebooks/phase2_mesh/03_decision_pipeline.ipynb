This notebook compares the single-agent baseline against the agentic mesh under the same model, task set, and run count.

## Decision framing

Use the mesh only when complexity and specialization gains outweigh coordination cost.

Under GCC-style constraints, the default bias should be toward the single agent for simple or latency-sensitive work.
Prefer the mesh only for multi-step, tool-heavy tasks where it improves success rate enough to justify higher token and coordination overhead.
