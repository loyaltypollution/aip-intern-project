# Phase 1: Single-Agent Baseline

## What this phase does
Runs a single LangGraph pipeline (triage → brief → response) 20 times on the
same citizen feedback inputs. Records latency, token cost, error rate, and
step trace per run to establish a control group.

## Architecture

```
START → triage_node → brief_node → response_node → END
```
All nodes use MCP filesystem tools to read from `workspace/data/` and write
outputs to `artifacts/{run_id}/outputs/`.

## How to run
See `01_run.ipynb` — load config, run sweep, inspect outputs.

## What to look for
- Is latency stable across 20 runs (low variance)?
- Are token counts consistent?
- Do all 3 output files get produced every run?

## Handoff to Phase 2
Phase 2 (mesh) uses the same workspace inputs and the same metrics schema.
Use `02_analysis.ipynb` to compare Phase 1 and Phase 2 results side by side.
